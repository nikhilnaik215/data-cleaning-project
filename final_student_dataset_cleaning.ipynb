```python
import pandas as pd
```

```python
df = pd.read_csv('student_dataset_v1.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns=['ListofCourses'])
```

```python
df
```

```python
df['ListofCourses'] = df['ListofCourses'].str.strip('/._ ') #Clean the text in this column by removing extra symbols
```

```python
df
```

```python
df = df.dropna(subset=['Gender'])       #remove rows with missing (NaN) values
df
```

```python
df.dtypes
```

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df
```

```python
print(df.head()) 
df.to_csv("cleaned_student_dataset.csv", index=False)
```
